```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.document_loaders import TextLoader

loader = PyPDFLoader('../resources/Insights.pdf')
textLoader = TextLoader("../resources/quotes.txt")
textFile = textLoader.load()
docs = loader.load()

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)


splits = text_splitter.split_documents(docs)
text_splits = text_splitter.split_documents(textFile)
print([doc.page_content for doc in text_splits])
print([doc.page_content for doc in splits])

```

```python
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.document_loaders import TextLoader

textLoader = TextLoader("../resources/quotes.txt")
textFile = textLoader.load()

splitter = CharacterTextSplitter( separator="\n\n", chunk_size=500, chunk_overlap=50)

splits = splitter.split_documents(textFile)
print([doc.page_content for doc in splits])
```

```python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_string = """
<div>
<h1>Hello World</h1>
<h2>Hello World Again</h2>
<p>Welcome to the world of Langchain</p>
</div>
"""

html_splitter = HTMLHeaderTextSplitter(headers_to_split_on=[("h1", "Header 1"), ("h2", "Header 2")])
html_splits = html_splitter.split_text(html_string)
print( html_splits)
```

```python
from langchain_text_splitters import RecursiveJsonSplitter
import json

json_data = [
    {
        "id": 1,
        "name": "Akshay Patil",
        "email": "akshay@example.com",
        "isActive": True,
        "roles": ["admin", "editor"]
    },
    {
        "id": 2,
        "name": "Jane Doe",
        "email": "jane@example.com",
        "isActive": False,
        "roles": ["viewer"]
    }
]

splitter = RecursiveJsonSplitter(max_chunk_size=30)

# Use split_json for a Python object; no need to convert to string
try:
    split_docs = splitter.split_json(json_data)
    print(split_docs)
except IndexError:
    # Handle the case where splitting produces no chunks
    print("No chunks produced (IndexError: list index out of range)")
```
