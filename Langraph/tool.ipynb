```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict

# 1. Define a schema for your data (even if it's empty)
class SimpleState(TypedDict):
    content: str

# 2. Initialize using StateGraph instead of Graph
workflow = StateGraph(SimpleState)

print("Import successful! StateGraph initialized.")
```

```python
def llm(input) -> str:
    from langchain_openai import ChatOpenAI

    gptLlm = ChatOpenAI(model_name="o3-mini")

    return gptLlm.invoke(input).content
```

```python
def token_counter(input) -> str:
    return f"The output token for the query are {len(input)}"
```

```python
workflow.add_node("func1", llm)
workflow.add_node("func2", token_counter)
workflow.add_edge("func1","func2")
workflow.set_entry_point("func1")
workflow.set_finish_point("func2")
workflow.run("what is capital of finland?")
```
