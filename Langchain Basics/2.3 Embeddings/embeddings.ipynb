```python
from langchain_openai import OpenAIEmbeddings

vectorizer = OpenAIEmbeddings(model="text-embedding-3-small", dimensions=256)

text = "Hello, I am a student."
vector = vectorizer.embed_query(text)
print(vector)
```

```python
from langchain_openai import OpenAIEmbeddings

vectorizer = OpenAIEmbeddings(model="text-embedding-3-small", dimensions=256)

from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter


quotesLoader = TextLoader("../../resources/quotes.txt")
quotesFile = quotesLoader.load()
splitter = CharacterTextSplitter( separator="\n\n", chunk_size=500, chunk_overlap=50)
splits = splitter.split_documents(quotesFile)

vectors = vectorizer.embed_documents([doc.page_content for doc in splits])
vectors

```
